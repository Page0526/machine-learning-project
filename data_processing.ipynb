```python
import pandas as pd
```

```python
df = pd.read_csv('movies_info.csv')
df
```

```python
#xoá hàng đầu tiên 
df = df.drop(0)
df.head()
```

```python
#các ô trống trong data
df.isnull().sum()
```

```python
#drop doanh thu domestic
df = df.drop(df.columns[0], axis=1)
```

```python
#tách earliest release data thành ngày ra mắt và nơi ra mắt đầu tiên
# Tách thành hai cột earlies release date và earliest release region
df[['Earliest Release date', 'Earliest Release region']] = df['Earliest Release Date'].str.split('(', expand=True)

# Loại bỏ dấu ngoặc đơn ở cuối của cột 'region'
df['Earliest Release region'] = df['Earliest Release region'].str.rstrip(')')

# Loại bỏ dấu cách thừa ở đầu và cuối của cả hai cột
df['Earliest Release date'] = df['Earliest Release date'].str.strip()
df['Earliest Release region'] = df['Earliest Release region'].str.strip()

# Hiển thị DataFrame sau khi tách
df = df.drop('Earliest Release Date', axis=1)
# Di chuyển cột mới tạo đến thứ 2 và 3
col_date = df.columns[8]
col_reg = df.columns[9]
df.insert(2, col_date, df.pop(col_date))
df.insert(3, col_reg, df.pop(col_reg))
#in ra df
df
```

```python
df.isnull().sum()
```

```python
#lấy các phim có mục Genres (thể loại) không null
df = df[df['Genres'].notnull()]
df
```

```python
#lấy các phim có mục Overview, Actors không null
df = df[(df['Overview'].notnull()) & (df['Actors'].notnull())]
df
```

```python
df.isnull().sum()
```

```python
#lấy các phim có mục Domestic Distribution (nhà phát hành) không null
df = df[df['Domestic Distributor'].notnull()]
df
```

```python
df.isnull().sum()
```

```python
# Thay các giá trị null trong trường 'MPAA' thành 'G' - phù hợp mọi lứa tuổi
df['MPAA'].fillna('G', inplace=True)
df.isnull().sum()
```

```python
# Thay các giá trị null trong trường 'Earliest Release region' thành 'domestic'
df.loc[:,'Earliest Release region'].fillna('domestic', inplace=True)
df
```

```python
df.isnull().sum()
```

```python
#chuyển tên các phim thành 1 list
list_name = df['Name'].tolist()
```

```python

#khởi tạo dataframe chưa doanh thu của 773 film trên
film_revenue = pd.DataFrame()

# Đọc file CSV vào DataFrame
df_1 = pd.read_csv('data.csv')
# Lọc các dòng chứa tên film từ tập hợp parent_asin_set
film_revenue = df_1[df_1['Release Group'].isin(list_name)]

# In ra DataFrame
film_revenue
```

```python
#lọc các trường cần cho output
film_revenue = film_revenue.loc[:,['Rank','Release Group','Worldwide','Domestic']]
film_revenue
```

```python
list = film_revenue['Release Group'].tolist()
print(len(list))
```

```python
# Lọc các dòng chứa tên film từ tập hợp list
film_revenue_final = pd.DataFrame()
film_revenue_final = df[df['Name'].isin(list)]
```

```python
film_revenue_final.shape[0]
```

```python
film_revenue_final
```

```python
# Gộp dataframe dựa trên cột chung là 'Name' và 'Release Group'
merged_df = pd.merge(film_revenue, film_revenue_final, how='inner', left_on='Release Group', right_on='Name')

# Loại bỏ cột 'Release Group' sau khi gộp
merged_df.drop(columns=['Release Group'], inplace=True)

merged_df
```

```python
merged_df['Genres'].dtype
```

```python
merged_df['Genres'] = merged_df['Genres'].apply(lambda x : x.split())
merged_df['Genres']
```

```python
merged_df['Genres'] = merged_df['Genres'].astype(str)
merged_df['Genres']
```

```python
merged_df.dtypes
```

```python
merged_df['Rank'] = merged_df['Rank'].astype('int64')
merged_df.head()
```

```python
#chuyển doanh thu worldwide về dạng float
merged_df['Worldwide'] = merged_df['Worldwide'].str.replace('$', '')
merged_df['Worldwide'] = merged_df['Worldwide'].str.replace(',','').astype(float)
merged_df['Worldwide']
```

```python
#chuyển doanh thu domestic về dạng float
merged_df['Domestic'] = merged_df['Domestic'].str.replace('$', '')
merged_df['Domestic'] = merged_df['Domestic'].str.replace(',','')
merged_df['Domestic']
```

```python
merged_df.to_csv('preprocessed_data.csv')
```
