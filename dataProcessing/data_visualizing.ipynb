```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
cd = os.getcwd()
parent_dir = os.path.dirname(cd)
data_path = os.path.join(parent_dir, 'data')
print(data_path)
```

```python
data = pd.read_csv(os.path.join(data_path, 'final_data.csv'), index_col=None)
```

```python
data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
data
```

## Top 10 highest-revenue movies all of year (from 2000-2024)

```python
top_worldwide_revenue = data.sort_values(by=['Worldwide'], ascending=False)[0:10]
top_worldwide_revenue
```

```python
plt.figure(figsize=(12, 6))
plt.plot(top_worldwide_revenue['Release Group'], top_worldwide_revenue['Worldwide'])
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.xlabel('Movies')
plt.ylabel('Worldwide Revenue (100 Millions dollar)')
plt.tight_layout()
plt.show()
```
